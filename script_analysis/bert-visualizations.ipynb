```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from math import pi # For radar chart
import os # To create an output directory
import sys # To use sys.exit() which is more reliable than exit() in some environments

# --- Create Directory for Plots ---
output_dir = 'plots'
os.makedirs(output_dir, exist_ok=True)
print(f"Plots will be saved in the '{output_dir}/' directory.")

# --- Configuration ---
# *** IMPORTANT: Verify this filename is correct ***
DATA_FILE = 'dialogues_with_vader_and_emotion.csv' # Make sure this matches your actual file name

# Define columns known NOT to be emotion scores
KNOWN_NON_EMOTION_COLS = ['film', 'character', 'dialogue', 'cleaned_dialogue', 'sentiment_score', 'sentiment_category', 'dominant_emotion']


# --- Load Data ---
print(f"Attempting to load data from: {DATA_FILE}")
try:
    df = pd.read_csv(DATA_FILE)
    print(f"Successfully loaded DataFrame. Shape: {df.shape}")
    print(f"Columns found: {df.columns.tolist()}")

    if df.empty:
         raise ValueError("Loaded DataFrame is empty.")

except FileNotFoundError:
    print(f"Error: Data file not found at '{DATA_FILE}'")
    print("Please ensure the file exists in the correct directory and the filename is correct.")
    sys.exit(1) # Exit the script cleanly
except Exception as e:
    print(f"An error occurred during data loading: {e}")
    sys.exit(1)

# --- Identify Emotion Columns ---
print("\nIdentifying emotion score columns...")
potential_emotion_columns = []
for col in df.columns:
    # Check if it's NOT a known non-emotion column
    if col not in KNOWN_NON_EMOTION_COLS:
        # Check if the column dtype is numeric (more robust check)
        if pd.api.types.is_numeric_dtype(df[col]):
            potential_emotion_columns.append(col)
        else:
            print(f"  - Skipping non-numeric column: '{col}' (dtype: {df[col].dtype})")

emotion_columns = potential_emotion_columns # Assign the identified columns

if not emotion_columns:
    print("\nError: Could not automatically identify any numeric emotion score columns.")
    print("Please check:")
    print("  1. If the expected emotion columns exist in the CSV.")
    print(f"  2. If their names are different from the KNOWN_NON_EMOTION_COLS list: {KNOWN_NON_EMOTION_COLS}")
    print("  3. If the emotion columns contain numeric data (not strings or objects).")
    sys.exit(1) # Exit if no emotion columns found

print(f"Successfully identified emotion columns: {emotion_columns}")


# --- Define Consistent Colors and Settings ---
print("\nSetting up plot styles and colors...")
try:
    sns.set_theme(style="whitegrid")
    # This line should now work because emotion_columns is guaranteed to exist and be non-empty
    emotion_palette = sns.color_palette("Set2", len(emotion_columns))
    emotion_color_map = dict(zip(emotion_columns, emotion_palette))
    film_palette = {"Pulp Fiction": "#1f77b4", "Goodfellas": "#ff7f0e"}
    print("Plot styles and colors defined.")
except Exception as e:
    print(f"An error occurred during plot setup: {e}")
    sys.exit(1)


# --- DATA IS LOADED AND VALIDATED - CONTINUE WITH PLOTTING ---

# --- Plot 1: Average Emotion Intensity per Character (Grouped Bar Chart) ---
print("\n--- Data for Plot 1: Average Emotion Scores per Character ---")
avg_emotions = df.groupby(['film', 'character'])[emotion_columns].mean()
# Add reset_index to make film/character regular columns for printing
print(avg_emotions.reset_index().to_markdown(index=False)) # Print as markdown table
# Melt for plotting
avg_emotions_melted = avg_emotions.reset_index().melt(id_vars=['film', 'character'],
                                                       var_name='emotion',
                                                       value_name='average_score')

print("\nGenerating & Saving Plot 1: Average Emotion Intensity per Character...")
plt.figure(figsize=(15, 8))
sns.barplot(data=avg_emotions_melted, x='character', y='average_score', hue='emotion', palette=emotion_color_map, dodge=True)
plt.title('Average Emotion Intensity per Character', fontsize=18, pad=20)
plt.xlabel('Character', fontsize=12)
plt.ylabel('Average Predicted Score', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Emotion', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.suptitle('Comparison Across Both Films', fontsize=14, y=0.95)
plt.tight_layout(rect=[0, 0, 0.85, 0.95])
plt.savefig(os.path.join(output_dir, 'plot1_avg_emotion_intensity_character.png'), dpi=300, bbox_inches='tight')
plt.show()


# --- Plot 2: Film Emotional Profiles (Radar Chart) ---
print("\n--- Data for Plot 2: Average Emotion Scores per Film ---")
film_avg_emotions = df.groupby('film')[emotion_columns].mean()
print(film_avg_emotions.reset_index().to_markdown(index=False)) # Print as markdown table

print("\nGenerating & Saving Plot 2: Film Emotional Profiles (Radar Chart)...")
labels = film_avg_emotions.columns.tolist()
num_vars = len(labels)
angles = [n / float(num_vars) * 2 * pi for n in range(num_vars)]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
for i, (film_name, row) in enumerate(film_avg_emotions.iterrows()):
    data = row.tolist()
    data += data[:1]
    ax.plot(angles, data, linewidth=2, linestyle='solid', label=film_name, color=film_palette[film_name])
    ax.fill(angles, data, color=film_palette[film_name], alpha=0.4)

ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)
max_score_radar = film_avg_emotions.max().max()
yticks = np.linspace(0, max_score_radar * 1.1, 5)
ax.set_yticks(yticks)
ax.set_yticklabels([f"{tick:.2f}" for tick in yticks], color="grey", size=8)
ax.set_ylim(0, max_score_radar * 1.1)
plt.title('Overall Emotional Profile by Film', size=20, y=1.1)
ax.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
plt.savefig(os.path.join(output_dir, 'plot2_film_emotion_profile_radar.png'), dpi=300, bbox_inches='tight')
plt.show()


# --- Plot 3 & 4: Dominant Emotion Analysis per Character (Stacked Bar Chart per Film) ---
print("\n--- Data for Plot 3 & 4: Percentage of Dominant Emotion per Character ---")
if 'dominant_emotion' not in df.columns:
     print("Calculating dominant emotion...")
     df['dominant_emotion'] = df[emotion_columns].idxmax(axis=1)

dominant_emotion_color_map = {emotion: color for emotion, color in zip(emotion_columns, emotion_palette)}
if 'neutral' in dominant_emotion_color_map:
    dominant_emotion_color_map['neutral'] = '#bdbdbd'

for film_name in df['film'].unique():
    print(f"\n--- Dominant Emotion Percentages for {film_name} ---")
    film_data = df[df['film'] == film_name].copy()
    char_dom_emotion_counts = film_data.groupby(['character', 'dominant_emotion']).size().unstack(fill_value=0)
    char_dom_emotion_counts = char_dom_emotion_counts.reindex(columns=emotion_columns, fill_value=0)
    char_dom_emotion_perc = char_dom_emotion_counts.apply(lambda x: x * 100 / x.sum() if x.sum() > 0 else 0, axis=1)

    # Print the percentage data for this film
    print(char_dom_emotion_perc.reset_index().to_markdown(index=False))

    print(f"\nGenerating & Saving Plot 3/4 for {film_name}...")
    plt.figure(figsize=(10, 7))
    ax = char_dom_emotion_perc.plot(kind='bar', stacked=True,
                                    color=[dominant_emotion_color_map.get(cat, '#000000') for cat in char_dom_emotion_counts.columns],
                                    width=0.8)
    safe_film_name = film_name.replace(' ', '_')
    plt.title(f'Dominant Emotion Expressed per Character\n({film_name})', fontsize=16)
    plt.xlabel('Character', fontsize=12)
    plt.ylabel('Percentage of Dialogue Lines (%)', fontsize=12)
    plt.xticks(rotation=0)
    plt.legend(title='Dominant Emotion', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.ylim(0, 100)
    plt.tight_layout(rect=[0, 0, 0.85, 1])
    plt.savefig(os.path.join(output_dir, f'plot3_dominant_emotion_{safe_film_name}.png'), dpi=300, bbox_inches='tight')
    plt.show()


# --- Plot 5: Emotion Correlation Heatmap ---
print("\n--- Data for Plot 5: Correlation Matrix ---")
if 'sentiment_score' in df.columns:
    corr_cols = ['sentiment_score'] + emotion_columns
    plot_title = 'Correlation Between VADER Sentiment and Predicted Emotions'
else:
    corr_cols = emotion_columns
    plot_title = 'Correlation Between Predicted Emotions'

corr_df = df[corr_cols]
correlation_matrix = corr_df.corr()
# Print the correlation matrix
print(correlation_matrix.to_markdown())

print("\nGenerating & Saving Plot 5: Emotion Correlation Heatmap...")
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title(plot_title, fontsize=16, pad=20)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig(os.path.join(output_dir, 'plot5_emotion_correlation_heatmap.png'), dpi=300, bbox_inches='tight')
plt.show()


# --- Plot 6: Emotion Distribution Comparison (Box Plot - Example: Anger) ---
KEY_EMOTION = 'anger' # Example: Change as needed
print(f"\n--- Data for Plot 6: Average '{KEY_EMOTION.capitalize()}' Scores per Character (for context) ---")

if KEY_EMOTION not in df.columns:
    print(f"Warning: Key emotion '{KEY_EMOTION}' not found in DataFrame columns. Skipping Plot 6.")
else:
    # Calculate and print mean scores for context
    avg_key_emotion = df.groupby(['film', 'character'])[KEY_EMOTION].mean().reset_index()
    print(avg_key_emotion.sort_values(KEY_EMOTION, ascending=False).to_markdown(index=False))

    print(f"\nGenerating & Saving Plot 6: Distribution of '{KEY_EMOTION.capitalize()}' Scores...")
    plt.figure(figsize=(12, 7))
    char_order = avg_key_emotion.sort_values(KEY_EMOTION, ascending=False)['character'].tolist() # Get order from printed data
    sns.boxplot(data=df, x='character', y=KEY_EMOTION, hue='film', palette=film_palette, order=char_order, showfliers=False)
    plt.title(f'Distribution of "{KEY_EMOTION.capitalize()}" Scores per Character', fontsize=16)
    plt.xlabel('Character (Ordered by Mean Score)', fontsize=12)
    plt.ylabel(f'Predicted "{KEY_EMOTION.capitalize()}" Score', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Film', loc='upper right')
    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, f'plot6_distribution_{KEY_EMOTION}.png'), dpi=300, bbox_inches='tight')
    plt.show()


# --- Qualitative Insights (No plot to save here) ---
print("\n--- Generating Data for Qualitative Insights ---")
print("Consider showing examples of high-scoring dialogues in your presentation.")

N_EXAMPLES = 5
for emotion in emotion_columns:
    if emotion in df.columns:
        print(f"\nTop {N_EXAMPLES} dialogues with highest '{emotion}' score:")
        top_dialogues = df.nlargest(N_EXAMPLES, emotion)[['film', 'character', 'cleaned_dialogue', emotion]]
        for index, row in top_dialogues.iterrows():
            dialogue_text = str(row['cleaned_dialogue'])
            print(f"  - {row['film']} | {row['character']} ({row[emotion]:.3f}): \"{dialogue_text[:100]}...\"")
    else:
        print(f"\nSkipping examples for missing emotion column: '{emotion}'")


print(f"\n--- Visualization and Insight Generation Complete ---")
print(f"All plots saved to the '{output_dir}/' directory.")
print("Numerical data for each plot was printed above the plot generation message.")
```
